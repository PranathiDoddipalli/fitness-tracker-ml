```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('/dailyActivity_merged_cleaned.csv')
```

Exploratory Data Analysis (EDA)

```python
data = data.set_index(data['ActivityDate'])
data.groupby(data['ActivityDate']).agg({'TotalSteps':'mean'}).plot(figsize=(15,5))
```

```python
data.groupby('TrackerDistance').agg({'Calories':'mean'}).plot(figsize=(15,5))
```

```python
data.groupby('TotalSteps').agg({'Calories':'mean'}).plot(figsize=(15,5))
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(data['TotalSteps'], data['Calories'], alpha=0.5)
plt.title('Total Steps vs. Calories Burned')
plt.xlabel('Total Steps')
plt.ylabel('Calories Burned')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(data['TotalDistance'], bins=30, color='green', alpha=0.7)
plt.title('Distribution of Total Distance')
plt.xlabel('Total Distance')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
activity_minutes = data.groupby(data['ActivityDate'])[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].mean()

plt.figure(figsize=(20, 5))
activity_minutes.plot(kind='line')
plt.title('Average Minutes Spent in Different Activity Levels per Activity Date')
plt.xlabel('Activity Date')
plt.ylabel('Average Minutes')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Activity Trends Over Time (example for VeryActiveMinutes)
plt.figure(figsize=(20,5))
sns.lineplot(x=data["ActivityDate"], y=data["VeryActiveMinutes"])
plt.xlabel("Activity Date")
plt.xticks(rotation=45)
plt.ylabel("Very Active Minutes")
plt.title("Very Active Minutes Over Time")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(data['SedentaryMinutes'], data['Calories'], alpha=0.5, color='orange')
plt.title('Sedentary Minutes vs. Calories Burned')
plt.xlabel('Sedentary Minutes')
plt.ylabel('Calories Burned')
plt.grid(True)
plt.show()
```

Sedentary minutes refer to the amount of time you spend sitting or lying down with minimal movement. It's distinct from light activity like walking slowly or doing chores.


```python
plt.figure(figsize=(15, 5))
plt.plot(data.groupby(data['ActivityDate'])['TotalDistance'].mean(), label='Average Total Distance', color='purple')
plt.plot(data.groupby(data['ActivityDate'])['TrackerDistance'].mean(), label='Average Tracker Distance', color='cyan')
plt.title('Trend of Total Distance and Tracker Distance Over Time')
plt.xlabel('Activity Date')
plt.ylabel('Distance (in miles)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
def categorize_activity(steps):
    if steps < 5000:
        return 'Sedentary'
    elif 5000 <= steps < 10000:
        return 'Lightly Active'
    elif 10000 <= steps < 15000:
        return 'Moderately Active'
    else:
        return 'Very Active'

data['ActivityLevel'] = data['TotalSteps'].apply(categorize_activity)

plt.figure(figsize=(10, 5))
data.boxplot(column='Calories', by='ActivityLevel', grid=True)
plt.title('Calories Burned by Activity Level')
plt.suptitle('')
plt.xlabel('Activity Level')
plt.ylabel('Calories Burned')
plt.show()
```

The boxplot comparison across activity levels shows that, generally, individuals in the 'Very Active' category tend to burn more calories, as evidenced by a higher median. However, there's a substantial overlap between the IQRs of different activity levels, especially between 'Moderately Active' and 'Very Active,' which suggests that the distinction between these groups might not always correlate with a significant difference in calories burned.

```python
#Check User
#Top 10 user with most calories burn
check_user = data.groupby('Id').agg({'Calories':'mean'}).sort_values(by='Calories', ascending=False).head(10)

plt.figure(figsize=(15,5))
sns.barplot(x=check_user.index, y=check_user['Calories'], order=check_user.index)
plt.xlabel('User ID')
plt.ylabel('Average Calories')
plt.title('Top 10 Users with Highest Average Calories')
plt.xticks(rotation=45)
plt.show()
```

```python
#Check User
#Top 10 user with most Steps
check_user = data.groupby('Id').agg({'TotalSteps':'mean'}).sort_values(by='TotalSteps', ascending=False).head(10)

plt.figure(figsize=(15,5))
sns.barplot(x=check_user.index, y=check_user['TotalSteps'], order=check_user.index)
plt.xlabel('User ID')
plt.ylabel('Average TotalSteps')
plt.title('Top 10 Users with Highest Average Total Steps')
plt.xticks(rotation=45)
plt.show()
```
